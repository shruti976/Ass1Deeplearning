```
w1=num.random.uniform(-2.0,2)
w2=num.random.uniform(-2.0,2)
w3=num.random.uniform(-2.0,2)
w4=num.random.uniform(-2.0,2)
w5=num.random.uniform(-2.0,2)
w6=num.random.uniform(-2.0,2)
w7=num.random.uniform(-2.0,2)
w8=num.random.uniform(-2.0,2)
w9=num.random.uniform(-2.0,2)
w10=num.random.uniform(-2.0,2)
w11=num.random.uniform(-2.0,2)
w12=num.random.uniform(-2.0,2)
w13=num.random.uniform(-2.0,2)
lr= 0.9
```

```
w14=num.random.uniform(-2.0,2)
w15=num.random.uniform(-2.0,2)
w16=num.random.uniform(-2.0,2)
w17=num.random.uniform(-2.0,2)
```

```
w = [[w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12],
     [w13,w14,w15,w16,w17]]
print("\n",w)
print("\nInitial weights(1st layer):\n")
for i in range(12):
    print("W%d: %f"% (i+1,w[0][i]))
print("\nInitial weights(2nd layer)\n")    
for i in range(5):
  print("W%d: %f"% (i+13, w[1][i]))
```

```
def sigmoid(x):
    return (1/(1+num.exp(-x)))
```

```
while True:
  total_error = 0 
  output = []
  for n in range(4):
      v1= (x1[n]*w1) + (x2[n]*w2) + w3 
      v2= (x1[n]*w4) + (x2[n]*w5) + w6
      v3= (x1[n]*w7) + (x2[n]*w8) + w9
      v4= (x1[n]*w10) + (x2[n]*w11) + w12

      
      y1=sigmoid(v1)
      y2=sigmoid(v2)
      y3=sigmoid(v3)
      y4=sigmoid(v4)
      #print(y1,y2,y3)

      u= y1*w13 + y2* w14 + y3*w15 + y4*w16 + w17

      y=sigmoid(u)
      output.insert(n,y)

      print(y)
      error = num.square(t[n]-y)
      d_error=t[n]-y

      dw13= lr* d_error * y * (1-y) * y1
      dw14= lr* d_error * y * (1-y) * y2
      dw15= lr* d_error * y * (1-y) * y3
      dw16= lr* d_error * y * (1-y) * y4
      dw17= lr* d_error * y * (1-y)
      

      w13 = dw13 + w13
      w14 = dw14 + w14
      w15 = dw15 + w15
      w16 = dw16 + w16
      w17 = dw17 + w17

      sum = lr * y*(1-y) * d_error * (w13 + w14 + w15 +w16 + w17)

      dw1 = y1* (1-y1) * x1[n] * sum
      dw2 = y1* (1-y1) * x2[n] * sum
      dw3 = y1* (1-y1) * sum
      dw4 = y2* (1-y2) * x1[n] * sum
      dw5 = y2* (1-y2) * x2[n] * sum
      dw6 = y2* (1-y2) * sum
      dw7 = y3* (1-y3) * x1[n] * sum
      dw8 = y3* (1-y3) * x2[n] * sum
      dw9 = y3* (1-y3) * sum
      dw10 = y4 * (1-y4) * x1[n] * sum
      dw11 = y4 * (1-y4) * x2[n] * sum
      dw12 = y4 * (1-y4) * sum 
      


      w1= w1 + dw1
      w2= w2 + dw2
      w3= w3 + dw3
      w4= w4 + dw4
      w5= w5 + dw5
      w6= w6 + dw6
      w7= w7 + dw7
      w8= w8 + dw8
      w9= w9 + dw9
      w10 = w10 + dw10
      w11 = w11 + dw11
      w12 = w12 + dw12
      w = [[w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12],
      [w13,w14,w15,w16,w17]]
      total_error = total_error+ error
      

  if(total_error<0.01):
    break
  

print("***********************************************************************************************************************************")
#print("\nInput data:", w)
print("\nTarget output", t)
print("\nPredicted output Y:", output)
print("\nWeights of first layer:")
for i in range(12):
  print("W%d: %f"% (i+1,w[0][i]))

print("\nWeights of second layer:")
for i in range(5):
  print("W%d: %f"% (i+13, w[1][i]))



  
```
